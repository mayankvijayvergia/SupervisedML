```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets  import  fetch_california_housing
```

```python
california = fetch_california_housing()
california
```

```python

california.keys()
```

```python
print(california.DESCR)
```

```python
california.feature_names
```

```python
#Let's prepare the dataset

df = pd.DataFrame(california.data, columns= california.feature_names)
df       # this all are independent columns

```

```python
# Now we have to add dependent column

df['Price'] =  california.target
df
```

```python
# Here More than one dependent value so here we use multiple linear regression

df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.corr()
```

```python
import seaborn as sns
sns.heatmap(df.corr(),annot=True)
```

```python
# Independent and dependent feature


X = df.iloc[:,:-1]   #independent dataset
y = df.iloc[:,-1]    #dependent dataset
```

```python
# SPlit the data into train and test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)  
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
print('The coefficent or slope :', regressor.coef_)
print('The intercept :',regressor.intercept_)
```

```python
y_pred_test = regressor.predict(X_test)
y_pred_test
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mse = mean_squared_error(y_test,y_pred_test)
mae = mean_absolute_error(y_test,y_pred_test)
rmse = np.sqrt(mse)

print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred_test)
score
```

```python
1 - (1-score)*(len(y_test)-1)/(len(y_test)- X_test.shape[1]-1)
```
