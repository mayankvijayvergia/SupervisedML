```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset.feature_names
```

```python
import pandas as pd

df_diabetes = pd.DataFrame(dataset.data,columns=dataset.feature_names)
df_diabetes
```

```python
X = df_diabetes
y = dataset['target']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

# Hyperparameter Tuning

```python
param = {
    'criterion' : ['squared_error', 'friedman_mse','absolute_error', 'poisson'],
    'splitter': ['best','random'],
    'max_depth' : [1,2,3,4,5],
    'max_features': ['auto','sqrt','log2']

}
```

```python
from sklearn.model_selection import GridSearchCV
treemodel = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(treemodel,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error


score = r2_score(y_test,y_pred)
print(score)
cm = mean_absolute_error(y_test,y_pred)
print(cm)
print(mean_squared_error(y_test,y_pred))
```

```python
# Now we dowith the selected MOdel

selectedModel = DecisionTreeRegressor(criterion='squared_error',max_depth=2,max_features='sqrt',splitter='best')
```

```python
selectedModel.fit(X_train,y_train)
```

```python
## Visualise The decision Tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectedModel,filled=True)
```
